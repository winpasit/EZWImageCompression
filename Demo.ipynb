```python
import numpy as np 
import matplotlib.pyplot as plt
import cv2
import os 
```

```python
# compress the images to ZTC file to decrease size of the picture
!python3 compress.py --output ./Demo/compressed_5.ztc --max-passes 5 ./Demo/thanit_handsome.png
!python3 compress.py --output ./Demo/compressed_10.ztc --max-passes 10 ./Demo/thanit_handsome.png
!python3 compress.py --output ./Demo/compressed_15.ztc --max-passes 15 ./Demo/thanit_handsome.png
!python3 compress.py --output ./Demo/compressed_20.ztc --max-passes 20 ./Demo/thanit_handsome.png
!python3 compress.py --output ./Demo/compressed_25.ztc --max-passes 25 ./Demo/thanit_handsome.png
!python3 compress.py --output ./Demo/compressed_inf.ztc --max-passes inf ./Demo/thanit_handsome.png
```

```python
compressed_filesize = {}
compressed_filesize['original'] = os.path.getsize('./Demo/thanit_handsome.png')
for file in ['compressed_5.ztc','compressed_10.ztc','compressed_15.ztc','compressed_20.ztc','compressed_25.ztc','compressed_inf.ztc']:
    compressed_filesize[file[:-4]] = os.path.getsize('./Demo/'+file)

compressed_filesize
```

```python
# decompress the images back to png file
!python3 decompress.py --output ./Demo/decompressed_5.png ./Demo/compressed_5.ztc
!python3 decompress.py --output ./Demo/decompressed_10.png ./Demo/compressed_10.ztc
!python3 decompress.py --output ./Demo/decompressed_15.png ./Demo/compressed_15.ztc
!python3 decompress.py --output ./Demo/decompressed_20.png ./Demo/compressed_20.ztc
!python3 decompress.py --output ./Demo/decompressed_25.png ./Demo/compressed_25.ztc
!python3 decompress.py --output ./Demo/decompressed_inf.png ./Demo/compressed_inf.ztc
```

```python
decompressed_filesize = {}
decompressed_filesize['original_img'] = os.path.getsize('./Demo/thanit_handsome.png')
for file in ['decompressed_5.png','decompressed_10.png','decompressed_15.png','decompressed_20.png','decompressed_25.png','decompressed_inf.png']:
    decompressed_filesize[file[:-4]] = os.path.getsize('./Demo/'+file)

decompressed_filesize
```

```python
original_img = cv2.imread('./Demo/Thanit_handsome.png')
decompressed_img_5 = cv2.imread('./Demo/decompressed_5.png')
decompressed_img_10 = cv2.imread('./Demo/decompressed_10.png')
decompressed_img_15 = cv2.imread('./Demo/decompressed_15.png')
decompressed_img_20 = cv2.imread('./Demo/decompressed_20.png')
decompressed_img_25 = cv2.imread('./Demo/decompressed_25.png')
decompressed_img_inf = cv2.imread('./Demo/decompressed_inf.png')

original_img = cv2.cvtColor(original_img, cv2.COLOR_BGR2RGB)
decompressed_img_5 = cv2.cvtColor(decompressed_img_5, cv2.COLOR_BGR2RGB)
decompressed_img_10 = cv2.cvtColor(decompressed_img_10, cv2.COLOR_BGR2RGB)
decompressed_img_15 = cv2.cvtColor(decompressed_img_15, cv2.COLOR_BGR2RGB)
decompressed_img_20 = cv2.cvtColor(decompressed_img_20, cv2.COLOR_BGR2RGB)
decompressed_img_25 = cv2.cvtColor(decompressed_img_25, cv2.COLOR_BGR2RGB)
decompressed_img_inf = cv2.cvtColor(decompressed_img_inf, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(20,10))
plt.subplot(2,4,1), plt.imshow(original_img), plt.title('Original: '+str(decompressed_filesize['original_img'])+' bytes')
plt.subplot(2,4,2), plt.imshow(decompressed_img_5), plt.title('decompressed_img_5: '+str(decompressed_filesize['decompressed_5'])+' bytes')
plt.subplot(2,4,3), plt.imshow(decompressed_img_10), plt.title('decompressed_img_10: '+str(decompressed_filesize['decompressed_10'])+' bytes')
plt.subplot(2,4,4), plt.imshow(decompressed_img_15), plt.title('decompressed_img_15: '+str(decompressed_filesize['decompressed_15'])+' bytes')
plt.subplot(2,4,6), plt.imshow(decompressed_img_20), plt.title('decompressed_img_20: '+str(decompressed_filesize['decompressed_20'])+' bytes')
plt.subplot(2,4,7), plt.imshow(decompressed_img_25), plt.title('decompressed_img_25: '+str(decompressed_filesize['decompressed_25'])+' bytes')
plt.subplot(2,4,8), plt.imshow(decompressed_img_inf), plt.title('decompressed_img_inf: '+str(decompressed_filesize['decompressed_inf'])+' bytes')
plt.show()
```
